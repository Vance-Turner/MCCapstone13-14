```python
import pandas as pd
import os
import numpy as np
```

```python
def loadCSVs(headLoss):
    RESUPath = os.path.join(os.getcwd(),'feb_4_bighl_1_ana','sim_'+str(headLoss)+'_CASE','RESU')
    resuDir = os.listdir(RESUPath)[0]
    monPath = os.path.join(RESUPath,resuDir,'monitoring')
    print "Loading:"+monPath
    dfx = pd.read_csv(os.path.join(monPath,'probes_VelocityX.csv'))
    dfy = pd.read_csv(os.path.join(monPath,'probes_VelocityY.csv'))
    dfz = pd.read_csv(os.path.join(monPath,'probes_VelocityZ.csv'))
    return (dfx,dfy,dfz)

def vel(df,*positions):
    return df[list(positions)][9:10]

def getMean(dfx,dfy,dfz,*positions):
    pos = list(positions)
    return (vel(dfx,pos[0],pos[1],pos[2],pos[3])**2 + vel(dfy,pos[0],pos[1],pos[2],pos[3])**2 + vel(dfz,pos[0],pos[1],pos[2],pos[3])**2).apply(np.sqrt).mean(axis=1)[9]

def getMean2(dfx,dfy,dfz,probe):
    """
    Get's the average velocity at a probe location
    """
    i = 1
    if probe==1:
        i = 1
    elif probe==2:
        i = 13
    elif probe==3:
        i = 25
    elif probe == 4:
        i = 37
    elif probe == 5:
        i = 49
    elif probe == 6:
        i = 65
    elif probe == 7:
        i = 81
    if probe < 5:
        inner = getMean(dfx,dfy,dfz,i,i+3,i+6,i+9)
        middle = getMean(dfx,dfy,dfz,i+1,i+1+3,i+1+6,i+1+9)
        outer = getMean(dfx,dfy,dfz,i+2,i+2+3,i+2+6,i+2+9)
        return (inner+middle+outer)/3.0
    else:
        inner = getMean(dfx,dfy,dfz,i,i+4,i+8,i+12)
        mid1 = getMean(dfx,dfy,dfz,i+1,i+1+4,i+1+8,i+1+12)
        mid2 = getMean(dfx,dfy,dfz,i+2,i+2+4,i+2+8,i+2+12)
        outer = getMean(dfx,dfy,dfz,i+3,i+3+4,i+3+8,i+3+12)      
        return (inner+mid1+mid2+outer)/4.0
```

```python
def plotCase(headLoss):
    (dfx,dfy,dfz) = loadCSVs(headLoss)
    col = ['t']
    for i in range(1,97):
        col.append(i)
    dfx.columns = col
    dfy.columns = col
    dfz.columns = col  
    means1=  []
    means2 = []
    means2_5 = []
    means3 = []
    labels = ['5','30','60','95','49','50.05','51']
    labels2 = ['49','50.05','51']
    for i in xrange(1,44,12):
        means1.append(getMean(dfx,dfy,dfz,i,i+3,i+6,i+9))
        means2.append(getMean(dfx,dfy,dfz,i+1,i+1+3,i+1+6,i+1+9))
        means3.append(getMean(dfx,dfy,dfz,i+2,i+2+3,i+2+6,i+2+9))
    for i in xrange(49,82,16):
        means1.append(getMean(dfx,dfy,dfz,i,i+4,i+8,i+12))
        means2.append(getMean(dfx,dfy,dfz,i+1,i+1+4,i+1+8,i+1+12))
        #means2_5.append(getMean(dfx,dfy,dfz,i+2,i+2+4,i+2+8,i+2+12))
        means3.append(getMean(dfx,dfy,dfz,i+3,i+3+4,i+3+8,i+3+12))
    print means1
    series1 = pd.Series(means1,labels)
    series2 = pd.Series(means2,labels)
    #series2_5 = pd.Series(means2_5,labels2)
    series3 = pd.Series(means3,labels)
    data = {'center':series1,'middle':series2,'outer':series3,}
    df = pd.DataFrame(data)
    ax = df.plot()
    ax.set_title('Velocities versus Position:'+str(headLoss))
    ax.figure.savefig('Vel_Pos_'+str(headLoss)+'.png')
```

```python
def plotCase2(headLoss,doPlotting=True):
    (dfx,dfy,dfz) = loadCSVs(headLoss)
    col = ['t']
    for i in range(1,97):
        col.append(i)
    dfx.columns = col
    dfy.columns = col
    dfz.columns = col 
    means = []
    labels = ['5','30','60','95','49','50.05','51']
    for i in xrange(1,8):
        means.append(getMean2(dfx,dfy,dfz,i))
    if doPlotting:
        pd.Series(means,labels).plot()
    else:
        return pd.Series(means,labels)
    
def plotCases(start,end,saveFigure=False):
    theSeries = {}
    data = {}
    for i in xrange(start,end):
        #theSeries.append(plotCase2(i/10.0,False))
        data['HL:'+str(i/10.0)] = plotCase2(i/10.0,False)
    df = pd.DataFrame(data)
    plt = df.plot()
    plt.set_title('Velocity versus Position:'+str(start/10.0)+'-'+str((end-1)/10.0))
    if saveFigure:
        plt.figure.savefig('VelPos_'+str(start/10.0)+'-'+str((end-1)/10.0)+'.png')
```
